```python
numberList = [1,12,3,4,15,6,7,8,19,10]
maxNumber = numberList[0]
for i in range(1,len(numberList)):
    if maxNumber <= numberList[i]:
        maxNumber = numberList[i]       
print(f"the maximum number in a list of integers is {maxNumber}")
```

# Question 18
# Write a for loop to find the common elements between two lists.

```python
list1 = [1,2,3,4,5]
list2 = [4,5,6,7,8]
commonList = []
for i in list1:
    for j in list2:
        if i == j:
            commonList.append(i)
print("common elements between two lists :",commonList)
```

# Question 19
# Write a for loop to remove all the duplicates from a list.

```python
list1 = [1, 2, 3, 14, 5, 3, 15, 8, 9, 2, 3,10,5]

# Initialize an empty list to store unique elements
unique_list = []

# Initialize a set to track seen elements
seen = set()

# Iterate over the list
for i in list1:
    # Check if the element has been seen before
    if i not in seen:
        # Add the element to the unique list
        unique_list.append(i)
        # Add the element to the seen set
        seen.add(i)

# Print the list without duplicates
print("List without duplicates:", unique_list)
```
