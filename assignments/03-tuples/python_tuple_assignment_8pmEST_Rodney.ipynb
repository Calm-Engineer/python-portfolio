1. What is a tuple in Python?

```python
row of strings or/and numbers and/or characters separated by commas stored in round brackets 
```

2. How do you create a tuple in Python? Provide an example.

```python
tuple1 = tuple(('apple','banana','cherry'))
tuple2 = tuple((1,2,3,4,5,6,7))
tuple3 = tuple(('apple','banana',3,4,5,6,7,'cherry',5,7,'#'))
print(tuple1)
print(tuple2)
print(tuple3)
```

3. How is a tuple different from a list in Python?

```python
# tuple
1. intergers/strings/special characters are stored in any combination inside a round bracket
2. can be indexed
3. immutable(unchangeable)
4. duplicated
5. ordered

# list
1. intergers or strings are stored in square brackets separately
2. can be indexed
3. mutable(changeable)
4. duplicated
5. ordered
```

4. Can a tuple contain elements of different data types? Provide an example.

```python
# Yes, tuple contains elements of different data types.
tuple3 = ('apple','banana',3,4,5,6,7,'cherry',5,7,'#')
print(tuple3)
print(type(tuple3[0]))
print(type(tuple3[3]))
print(type(tuple3[-1]))
```

16. What is tuple unpacking in Python? Provide an example.

```python
tuple1 = ('apple','banana','cherry')
(x1,x2,x3) = tuple1
print(x1)
print(x2)
print(x3)

# OR

fruits = ('apple','banana','cherry','strawberry')
(y1,*y2,y3) = fruits
print(y1)
print(y2)
print(y3)
```

19. Explain the immutability of tuples in Python.

```python
# Tuples are immutable(unchangeable): once the tuple has been created, we cannot change, add, or remove items
fruits = ('apple','banana','cherry','strawberry')
fruits[2]='mango'
print(fruits)
```
