15. Check if the list `[1, 2, 3]` is a subset of the list `[1, 2, 3, 4, 5]`.


```python
# Note: Used ChatGPT to solve this problem

# Method 1: Using sets
list1 = [1, 2, 3]
list2 = [1, 2, 3, 4, 5]

is_subset = all(i in list2 for i in list1)
if is_subset == True:
    print("list1 is a subset of list2")
else:
    print("list1 is NOT a subset of list2")

# OR 

# Method 2: Converting to sets and using issubset()
list1 = [1, 2, 3]
list2 = [1, 2, 3, 4, 5]

set1 = set(list1)
set2 = set(list2)
is_subset = set1.issubset(set2)
if is_subset == True:
    print("list1 is a subset of list2")
else:
    print("list1 is NOT a subset of list2")

```
