```python
# Question 3: Create a list with mixed data types and count the occurrences of a specific type.
# Create a list with mixed data types
original_list = [1, 2, 3, 'a', 'b']

# Initialize a counter for integers
count_int = 0

# Iterate through the list and count the occurrences of int type
for i in original_list:
    if isinstance(i, int):
        count_int += 1

# Initialize a counter for strings
count_str = 0

# Iterate through the list and count the occurrences of str type
for x in original_list:
    if isinstance(x, str):
        count_str += 1

# Print the count of integers
print(count_int)

# Print the count of strings
print(count_str)
```

```python
# Question 4: Count the occurrences of an element that is not in the list.
original_list = [1, 2, 3,4,5]

# Initialize a counter for strings
count_str = 0

# Iterate through the list and count the occurrences of str type
for x in original_list:
    if isinstance(x, str):
        count_str += 1

# Print the count of integers
print(count_str)

# Print the count of 7 in the list
print(original_list.count(7))
```

```python
# Question 5: Count the occurrences of an element in an empty list.
original_list = []

# Initialize a counter for strings
count_int = 0

# Iterate through the list and count the occurrences of str type
for x in original_list:
    if isinstance(x, int):
        count_int += 1

# Print the count of integers
print(count_int)

# Print the count of 7 in the list
print(original_list.count(7))
```
