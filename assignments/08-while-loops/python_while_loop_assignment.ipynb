# Question 18
# Write a while loop to find the common elements between two lists.

```python
list1 = [1,2,3,4,5]
list2 = [4,5,6,7,8]
commonList = []
i = 0
while i < len(list1):
    j = 0
    while j < len(list2):
        if list1[i] == list2[j]:
            commonList.append(list1[i])
        j += 1
    i += 1            
print("common elements between two lists :",commonList)
```

# Question 19
# Write a while loop to remove all the duplicates from a list.

```python
list1 = [1, 2, 13, 4, 15, 3, 5, 8, 9, 2,10 ]

# Initialize an empty list to store unique elements
unique_list = []

# Initialize a set to track seen elements
seen = set()

# Iterate over the list
i = 0
while i < len(list1):
    # Check if the element has been seen before
    if list1[i] not in seen:
        # Add the element to the unique list
        unique_list.append(list1[i])
        # Add the element to the seen set
        seen.add(list1[i])
    i += 1

# Print the list without duplicates
print("List without duplicates:", unique_list)
```

# Question 20
# Write a while loop to generate a list of the first 10 prime numbers.

```python
# Initialize an empty list to store prime numbers
prime_numbers = []

# Initialize a counter for the number of primes found
count = 0

# Start checking from the first prime number, which is 2
number = 2

# Loop until we have found 10 prime numbers
while count < 10:
    # Assume the number is prime until proven otherwise
    is_prime = True

    # Check if the number is divisible by any number less than itself
    i = 2
    while i < number:
        if number % i == 0:
            is_prime = False
            break
        i += 1

    # If the number is prime, add it to the list and increment the count
    if is_prime:
        prime_numbers.append(number)
        count += 1

    # Move to the next number
    number += 1

# Print the list of the first 10 prime numbers
print("The first 10 prime numbers are:", prime_numbers)
```
