# 8. Advanced Set Operations

```python
# 8.1 Write a program to find all subsets of a given set. (Got help from ChatGPT)
def get_all_subsets(input_set):
    # Convert the set to a list for easier manipulation
    input_list = list(input_set)
    n = len(input_list)
    
    # Initialize a list to hold all subsets
    all_subsets = []
    
    # There are 2^n possible subsets (including the empty set and the set itself)
    for i in range(2**n):
        subset = []
        for j in range(n):
            # Check if the j-th element is in the i-th subset
            if (i & (1 << j)) > 0:
                subset.append(input_list[j])
        all_subsets.append(subset)
    
    return all_subsets

# Given set
set1 = {'apple', 'banana', 'cherry'}

# Get all subsets
subsets = get_all_subsets(set1)

# Print all subsets
for subset in subsets:
    print(subset)
```

```python
# 8.2 Implement a function to compute the Cartesian product of two sets.(Got help from ChatGPT)
def cartesian_product(set1, set2):
    # Initialize an empty list to store the Cartesian product
    product = []
    
    # Iterate over each element in the first set
    for a in set1:
        # Iterate over each element in the second set
        for b in set2:
            # Append the ordered pair (a, b) to the product list
            product.append((a, b))
    
    return product

# Example usage:
set1 = {1, 2}
set2 = {'a', 'b'}

# Compute the Cartesian product
result = cartesian_product(set1, set2)

# Print the result
print(result)
```

```python
# 10.2 Write a program to measure the time taken to add elements to a set versus a list. (Got help from ChatGPT)
import time

# Number of elements to add
num_elements = 1000000

# Measure time to add elements to a set
start_time_set = time.time()
my_set = set()
for i in range(num_elements):
    my_set.add(i)
end_time_set = time.time()
time_taken_set = end_time_set - start_time_set

# Measure time to add elements to a list
start_time_list = time.time()
my_list = []
for i in range(num_elements):
    my_list.append(i)
end_time_list = time.time()
time_taken_list = end_time_list - start_time_list

# Print the results
print(f"Time taken to add {num_elements} elements to a set: {time_taken_set:.6f} seconds")
print(f"Time taken to add {num_elements} elements to a list: {time_taken_list:.6f} seconds")
```
