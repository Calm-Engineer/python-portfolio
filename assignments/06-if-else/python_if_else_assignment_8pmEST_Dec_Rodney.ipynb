# Question 12
# Write an if-else statement to check if a given number is a perfect square.

```python
# got help from ChatGPT
import math

# Given number
number = int(input("Enter a number: "))

# Check if the number is a perfect square
if number >= 0:
    root = math.sqrt(number)
    if int(root) ** 2 == number:
        print(f"{number} is a perfect square.")
    else:
        print(f"{number} is not a perfect square.")
else:
    print(f"{number} is not a perfect square.")
```

# Question 14
# Write an if-else statement to determine the grade based on a given score (A: 90-100, B: 80-89, C: 70-79, D: 60-69, F: <60).

```python
marks = int(input("Enter marks :"))
if 100 >= marks >= 90:
    grade = "A"
elif 89 >= marks >= 80:
    grade = "B"
elif 79 >= marks >= 70:
    grade = "C"
elif 69 >= marks >= 60:
    grade = "D"    
else:
    grade = "F"    
print(f"The grade based on the score {marks} is {grade} ")
```

# Write an if-else statement to check if two given strings are anagrams.
