12. Create a new dictionary ages_over_30 containing only the key-value pairs from
sample_dict where the age is over 30.

```python
# Sample dictionary
sample_dict = {'Alice': 25, 'Bob': 30, 'Charlie': 35, 'David': 40, 'Emily': 22}

# Initialize an empty dictionary to store the result
new_dict_over30 = {}

# Iterate over the key-value pairs in the sample dictionary
for key, value in sample_dict.items():
    if value > 30:
        new_dict_over30[key] = value

# Print the new dictionary
print(new_dict_over30)
```

17. Sort sample_dict by its keys and print the sorted dictionary.

```python
sample_dict = { 'Alice': 25, 'Bob': 30, 'Charlie': 35, 'David': 40, 'Emily': 22}
sorted_keys_sample_dict = {key: value for key, value in sorted(sample_dict.items(), key=lambda item: item[0])}
print(sorted_keys_sample_dict)
```

18. Sort sample_dict by its values and print the sorted dictionary.

```python
sample_dict = { 'Alice': 25, 'Bob': 30, 'Charlie': 35, 'David': 40, 'Emily': 22}

# Sort the dictionary by values and create a new dictionary
sorted_values_sample_dict = {key: value for key, value in sorted(sample_dict.items(), key=lambda item: item[1])}

# Print the sorted dictionary
print(sorted_values_sample_dict)
```
